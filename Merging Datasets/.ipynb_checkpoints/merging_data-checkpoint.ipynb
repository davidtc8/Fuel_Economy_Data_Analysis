# Merging Datasets
Use pandas Merges to create a combined dataset from `clean_08.csv` and `clean_18.csv`.

```python
# Merging the two dataframes together.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# load datasets
df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
df_08.head(2)
```

### Create combined dataset

```python
# rename 2008 columns
df_08.rename(columns={"model": "model_2008", 
                      "displ": "displ_2008", 
                      "cyl": "cyl_2008", 
                      "trans": "trans_2008",
                      "drive": "drive_2008",
                      "fuel": "fuel_2008",
                      "veh_class": "veh_class_2008",
                      "air_pollution_score": "air_pollution_score_2008",
                      "city_mpg": "city_mpg_2008",
                      "hwy_mpg": "hwy_mpg_2008",
                      "cmb_mpg": "cmb_mpg_2008",
                      "greenhouse_gas_score": "greenhouse_gas_score_2008",
                      "smartway": "smartway_2008"}, inplace = True)
```

```python
# view to check names
df_08.head()
```

```python
# merge datasets
df_combined = df_08.merge(df_18, left_on='model_2008', right_on='model', how='inner')
```
